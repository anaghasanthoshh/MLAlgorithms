```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

##**Loading the data:**

```python

pima=pd.read_csv('diabetes.csv',header=None)
pima.columns=pima.iloc[0]
pima=pima[1:]
```

```python
pima.head()
```

##**Feature Selection:**

```python
#split dataset in features and target variable
feature_cols = pima.columns[:-1]
X=pima.loc[:,pima.columns[:-1]]
y=pima['Outcome']#the target
X.head()
```

##**Splitting Data into Train and Test:**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,shuffle=True,random_state=1)
```

##**Building the Decision Tree Model:**

```python
dt=DecisionTreeClassifier(criterion='gini')
dt=dt.fit(X_train,y_train)

#prediction
pred=dt.predict(X_test)
```

##**Evaluating the model**

```python
accuracy=metrics.accuracy_score(y_test,pred)
print('Accuracy:',accuracy)
```

```python
from sklearn import tree
tree.plot_tree(dt);

```

##**Parameter Tuning:**

```python

dt1=DecisionTreeClassifier(criterion='gini',max_depth=3,splitter='best')
#Lets perform some parameter tuning:
dt1=DecisionTreeClassifier(criterion='gini',max_depth=3,splitter='best')
pred1=dt1.fit(X_train,y_train).predict(X_test)
accuracy1=metrics.accuracy_score(y_test,pred1)
print(f'Accuracy:{accuracy1:.2f}')
```
